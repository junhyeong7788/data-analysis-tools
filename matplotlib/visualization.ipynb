## 데이터 시각화
- 수집된 데이터에 대한 분석 결과를 쉽게 이해할 수 있도록 여러 시각화 도구를 통해 효과적으로 정보를 전달하여 의사결정 기회를 부여
- 데이터의 시각화는 문제를 정의하고 정의된 문제를 해결하는데 필요한 데이터를 수접하여 전처리 과정을 거쳐 데이터 분석 목적에 맞도록 모델링한 결과를 보여줌  

- 데이터 시각화 장점
    - 많은 양의 데이터를 시각적으로 요약
    - 시각을 통한 데이터 인사이트 도출
    - 더 정확한 데이터 분석 결과 도출

### 데이터 시각화의 종류
- 목적과 의도 및 데이터의 유형에 따라 구분 가능
    - 시간 시각화 : 시간의 변화에 따른 데이터 표현 (막대, 누적 막대, 점 그래프)
    - 관계 시각화 : 데이터들 사이에 존재하는 관계를 표현 (히스토그램, 버블차트, 박스플롯)
    - 비교 시각화 : 데이터의 값들을 비교하여 표현 (히트맵, 스타 차트, 평행 좌표계, 다차원 척도법)
    - 분포 시각화 : 전체의 관점에서 각 부분 간의 관계 표현 (파이 차트, 도우넛 차트, 트리맵, 누적 연속 그래프)
    - 공간 시각화 : 지도상에 해당하는 정보를 표현 (지도 맵핑)

##### X, Y축에 표기할 데이터 지정

```python
# 년도별 매출 현황
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022]
sales = [11865.0, 21865.0, 31865.0, 41865.0, 35865, 40865.3, 50865.3]

# 선 그래프를 그린다. x축에는 years값, y축에는 sales 값이 표시된다.
plt.plot(years, sales, color='green', marker='o', linestyle='solid')

# 제목을 설정한다.
plt.title("년도별 매출 현황")

# y축에 레이블을 붙인다.
plt.ylabel("sales")
# plt.savefig("sales.png", dpi=600)  # png 이미지로 저장 가능
plt.show()
```

```python
plt.plot(sales, years, color='red', marker='o', linestyle='solid')
```

##### 스타일 지정

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         color='m', linestyle=':', linewidth='5', marker='o',
         markersize='15', markeredgecolor='b', markeredgewidth='3')
plt.show()
```

##### ex
- 그래프 제목
- X = 10부터 시작해서 40까지 5씩 증가하는 정수, numpy함수를 이용하여 구현
- y = 7부터 시작하는 홀수. list comprehension으로 구현
- 스타일 지정

```python
plt.title("Setting ranges of x and y")

# 리스트 컴프리헨션 사용
plt.plot(np.linspace(10, 40, 7), [i for i in range(7, 20, 2)],
         c="b", lw=3, ls="--", marker="H", ms=15, mec="g", mew=3, mfc="c")

plt.xlim(0, 50)
plt.ylim(0, 30)
plt.grid()
plt.show()
```

##### 여러 선 차트 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("하나의 plot 함수 이용")
plt.plot(t, t, 'r--',
         t, 0.5 * t**2, 'bs:',
         t, 0.2 * t**3, 'g^-')
plt.show()
```

```python
x = [x for x in range(-20, 20)]  # -20에서 20 사이의 수를 1의 간격으로 생성
y1 = [2*t for t in x]            # 2*x를 원소로 가지는 y1 함수
y2 = [t**2 + 5 for t in x]       # x**2 + 5를 원소로 가지는 y2 함수
y3 = [-t**2 - 5 for t in x]      # -x**2 - 5를 원소로 가지는 y3 함수

# 빨강색 점선, 녹색 실선과 세모기호, 파랑색 별표와 점선으로 각각의 함수를 표현
plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.axis([-30, 30, -30, 30])     # 그림을 그릴 영역을 지정함
plt.show()
```

#### 막대 그래프 : 데이터 크기를 막대로 표현한 것, 각 데이터 집단 간 차이를 비교하기 좋음

```python
y = [1.8, 2.5, 0.7]
x = np.arange(len(y))
# print(x)
xlabel = ['A', 'B', 'C']
plt.title("Bar Chart")
plt.bar(x, y)

# 사용법 matplotlib.pyplot.xticks 검색
plt.xticks(x, xlabel) # 주석 처리 후 결과 비교하기
plt.yticks(y)
plt.xlabel("name")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.bar(x, y, width=0.4)
plt.show()
```

```python
# 연도별 매출 현황
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022]
sales = [11865.0, 21865.0, 31865.0, 41865.0, 35865.0, 40865.3, 50865.3]

plt.barh(years, sales)  # 가로 막대그래프
plt.title("연도별 매출 현황") # 제목을 설정한다.
plt.ylabel("sales")      # y축에 레이블을 붙인다.
plt.show()
```

```python
plt.bar(range(len(years)), sales)

plt.title("년도별 매출 현황")
plt.ylabel('sales')

plt.xticks(range(len(years)), years)
plt.show()
```

#### 누적 막대 그래프

```python
a = [3, 20, 40, 31]
b = [11, 22, 55, 33]

plt.bar(np.arange(4), a)
plt.bar(np.arange(4), b, bottom=a)
plt.show()
```

```python
# 누적 막대 그래프
x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)

data = [x, y, z]
x_array = np.arange(10)

# 누적 막대 그래프 그리기
for i in range(0, 3):
    plt.bar(
        x_array,
        data[i],
        # 누적하기 위해 bottom 업데이트
        bottom=np.sum(data[:i], axis=0)
    )
plt.show()
```
